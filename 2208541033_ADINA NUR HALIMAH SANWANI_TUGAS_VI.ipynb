5.21.4 MENGUBAH DATA PADA PANDAS

```python
# mengubah data pada kolom dataframe sepal length (cm) menjadi (mm)

data1['sepal length (mm)'] = 10*data1['sepal length (cm)']
print(data1)
# data yang dimasukkan akan diinput sebagai kolom terbaru dan berada di index kolom terakhir
```

```python
# penggunaan drop untuk menghapus data dalam dataframe

data1 = data1.drop(columns = ['sepal length (mm)'])
print(data1)
# pada code ini akan dihapus kolom sepal length (mm)
```

```python
# menggunakan operasi penjumlahan pada pengubahan data

data1['Jumlah panjang dan lebar daun bunga'] = data1['petal length (cm)']+data1['petal width (cm)']
print(data1)
# data yang dimasukkan akan terinput sebagai kolom terbaru dan berada di index terakhir
```

```python
data1 = data1.drop(columns=['Jumlah panjang dan lebar daun bunga'])
print(data1)

# menghapus semua data kolom Jumlah panjang dan lebar daun bunga
```

```python
# cara lain untuk melakukan operasi penjumlahan pada pandas

data1['Jumlah panjang dan lebar daun bunga'] = data1.iloc[:,2:4].sum(axis=1)
print(data1)
# iloc berfungsi untuk menampilkan posisi
# : di depan koma berperan sebagai baris, karena tidak ada argumen/integer maka semua data dalam dataframe bunga iris akan ditampilkan
# 2:4 di belakang koma berperan sebagai kolom sehingga akan dibaca yaitu kolom index nomor 2 dan kolom index nomor 3
# sum(axis=1) berfungsi untuk menjumlahkan nilai pada kolom index nomor 3 dan kolom index nomor 4 dengan kolom+kolom
```

```python
# mengganti data dalam kolom dataframe

data1['sepal width (cm)'] = data1['sepal width (cm)'].replace([3.0], 2.2)
print(data1)
# diganti data dalam sepal width (cm) yang bernilai 3.0 menjadi 2.2
```

5.22.2 PROPERTI PEMBUATAN GRAFIK DENGAN MATPLOTLIB

```python
# pemanggilan data pada bunga iris
data2 = pd.read_csv('bunga_iris.csv')

x = data2['sepal length (cm)']
y = data2['petal length (cm)']
x1 = data2['sepal width (cm)']
y1 = data2['petal width (cm)']

plt.figure(figsize=(16,20), dpi=400) 
# figure merupakan keseluruhan visualisasi yang kita plot dalam 1 code
# figsize berfungsi untuk mengatur ukuran grafik
# dpi merupakan resolusi gambar

plt.plot(x,y, label = 'Garis Length',color='lightpink',linewidth =2,marker='*',markersize=15,markeredgecolor='red',linestyle='-')
plt.plot(x1,y1, label = 'Garis Width',color='red',linewidth=2,marker='*',markersize=15,markeredgecolor='pink',linestyle='-')
# label untuk memberi nama label pada legend
# line width merupakan ketebalan garis
# marker berfungsi untuk mengubah tipe point data di chart
# markersize berfungsi untuk mengubah ukuran pada marker
# markeredgecolor berfungsi untuk memberikan warna pada pinggiran marker
# linestyle berfungsi untuk mengubah jenis dari garis

plt.title('Grafik Bunga Iris', fontdict = {'fontname' : 'Comic Sans MS' , 'fontsize' : 14}) #memberi nama pada judul grafik
plt.xlabel('X', fontdict={'fontname' : 'Times New Roman' , 'fontsize' : 12}) #memberi nama pada sumbu x
plt.ylabel('Y', fontdict={'fontname' : 'Times New Roman' , 'fontsize' : 12}) #memberi nama pada sumbu y
# fontname berfungsi untuk mengubah judul dengan font yang dipilih
# fontsize berfungsi untuk mengubah besar/kecilnya font

plt.xticks(data2['sepal length (cm)'])
plt.yticks(data2['petal length (cm)'])
plt.legend(loc='upper center')

plt.xticks(data2['sepal width (cm)'])
plt.yticks(data2['petal width (cm)'])
plt.legend(loc='upper center')
# ticks merupakan penanda berupa titik yang berjajar di sumbu, sebagai refrensi skala nilai
# legend merupakan label data yang muncul dalam plot

plt.savefig('grafik.png', dpi=400)
# savefig berfungsi untuk menyimpan grafik pada folder, dengan resolusi 400 (menyesuaikan dpi)
plt.show()
```

```python
# dimasukkan data x, y, x1 dan y1 sebagai berikut
x = [-3,-2,-1,0,1,2,3]
y = [27,8,1,0,1,8,27]
x1 = [-3,-2,-1.5,-1.3,-1.1,0,1.2,1.7,2.2,2.7]
y1 = [-1,1,4,7,8,10,17,21,22,25]


plt.figure(figsize=(4,5),dpi=400)
# figure merupakan keseluruhan visualisasi yang kita plot dalam 1 code
# figsize berfungsi untuk mengatur ukuran grafik
# pada code ini figure memiliki ukuran 5x4
# dpi merupakan resolusi gambar

plt.plot(x,y,label='y=x^3',color='yellow',linewidth=2,marker='*',markersize=10, markeredgecolor='red',linestyle='-')
# label untuk memberi label pada legend
# line width merupakan ketebalan garis
# marker berfungsi untuk mengubah tipe point data di chart. bisa menggunakan bentuk lain seperti 'o','x','s [persegi]','D [wajik]'
# markersize berfungsi untuk mengubah ukuran pada marker
# markeredgecolor berfungsi untuk memberikan warna pada pinggiran marker
# linestyle berfungsi untuk mengubah jenis dari garis. bentuk lain linestyle adalah '-',':','-.','.'

plt.plot(x1,y1)
plt.title('Grafik y=x^3',fontdict={'fontname':'Comic Sans MS','fontsize':16})
# menambahkan judul pada grafik
# fontdict untuk mengubah font pada judul grafik
# fontname berfungsi untuk mengubah judul dengan font yang dipilih
# fontsize berfungsi untuk mengubah besar/kecilnya font

plt.xlabel('x',fontdict={'fontname':'Calibri','fontsize': 14})
plt.ylabel('y',fontdict={'fontname':'Calibri','fontsize':14})
plt.xticks([-3,-2,-1,0,1,2,3])
# ticks merupakan penanda berupa titik yang berjajar di sumbu, sebagai refrensi skala nilai
plt.yticks([27,8,1,0,1,8,27])
plt.legend(loc='upper center')
# legend merupakan label data yang muncul dalam plot
plt.savefig('grafik.png',dpi=400)
# savefig berfungsi untuk menyimpan grafik pada folder, dengan resolusi 400 (menyesuaikan dpi)
plt.show()
```

5.22.4 PROPERTI PEMBUATAN HISTOGRAM DENGAN MATPLOTLIB

```python
# dimasukkan data sebagai berikut
penyanyi = ['Jungkook','RM','Taehyung','Jin']
penggemar = [6000,4700,1760,5000]

plt.figure(figsize=(9,4))
# menghasilkan gambar ukuran 9x4
histogram = plt.bar(penyanyi,penggemar,color='lightpink')
# penggunaan color untuk mengganti warna pada bar

# hatch berfungsi untuk membuat arsiran pada bar
histogram[0].set_hatch('x') #dipilih histogram dengan index nomor 0 berarsir x
histogram[1].set_hatch('o')
histogram[2].set_hatch('.')
histogram[3].set_hatch('/')
plt.show()
```

```python
penyanyi = ['Jungkook','RM','Taehyung','Jin']
penggemar = [6000,4700,1760,5000]

plt.figure(figsize=(9,4))
# menghasilkan gambar ukuran 9x4
histogram = plt.bar(penyanyi,penggemar,color='lightgreen')
pola = ['*','/','o','.']

# penggunaan for
for i in [0,1,2,3]: #pada i dengan index nomor 0,1,2,3
    histogram[i].set_hatch(pola[i]) #akan dicetak arsiran pada varibel pola sesuai nomor index
    
plt.show()
```
